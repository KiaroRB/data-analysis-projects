```python
#Columnas tipo object
cols_object = df.select_dtypes(include='object').columns

#Reemplazar comas y símbolos solo en columnas string
for col in cols_object:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace(',', '', regex=False)
    df[col] = df[col].str.replace('$', '', regex=False)
    df[col] = df[col].str.strip()

#Convertir a número cuando sea posible
df[cols_object] = df[cols_object].apply(
    lambda x: pd.to_numeric(x, errors='ignore')
)

df.dtypes

```

```python
df.info()
```

```python
df = df.drop(columns=['TIDAL Popularity'])
```

```python
df.select_dtypes(include='object')
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'], errors ='coerce')
```

```python
df.info()
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
obj_cols = df.select_dtypes(include=["object"]).columns
df[obj_cols] = df[obj_cols].fillna("Desconocido")
```
